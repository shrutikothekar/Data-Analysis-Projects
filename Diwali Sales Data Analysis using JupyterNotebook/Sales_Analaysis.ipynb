# total amount vs age group 

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=sales_age, x= "Age Group", y= "Amount")
```

# State work

```python
df.columns
```

# Total numbers of orders from top 10 states 

```python
sales_states = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by="Orders", ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})    #  (width,height)
sns.barplot(data=sales_states, x="State",y="Orders")
```

# Total numbers of amounts from top 10 states 

```python
sales_states = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})    #  (width,height)
sns.barplot(data=sales_states, x="State",y="Amount")
```

# FROM ABOVE GIVEN GRAPH WE CAN SEE THAT MOST OF ORDERS ARE FROM UTTER PRADESH , MAHARASHTRA , KARNATAKA RESPESCTIVY. 

# Marital Status

```python
status_count = sns.countplot(data=df, x="Status")
sns.set(rc={'figure.figsize':(7,5)})    #  (width,height), , hue="Status"
for i in status_count.containers:
    status_count.bar_label(i)
```

```python
status_sales = df.groupby(['Status','Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={'figure.figsize':(7,5)})    #  (width,height), , hue="Status"
sns.barplot(x="Status", y="Amount",data=status_sales,hue='Gender')
```

# FROM ABOVE GIVEN GRAPH WE CAN SEE THAT MOST OF THE BUYERS ARE MARRIER WOMEN ANS THEY HAVE HIGH 
# PURCHESSING POWER

# Product Category

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(25,5)})
# count product category wise
count_product = sns.countplot(x="Product_Category", data=df)
for i in count_product.containers:
    count_product.bar_label(i)
```

```python
sales_product = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False).head(12)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data = sales_product, x='Product_Category', y='Amount')
```

# FRON THE ABOVE GIVEN GRAPH, WE CAN SEE THAT MOST OF SOLD PRODUCT ARE FROM FOOD, CLOTHINGS, AND ELECTRONICS

```python
sales_proid = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(15)
sns.set(rc={'figure.figsize':[20,5]})
sns.barplot(data=sales_proid, x="Product_ID", y="Orders")
```
